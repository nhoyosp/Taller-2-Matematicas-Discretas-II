**Punto 2.3.a**

Podemos aprovechar que $\frac{1}{t_k}=\frac{1}{\frac{k\cdot (k+1)}{2}}=\frac{2}{k\cdot (k+1)}$ por fracciones parciales $2*(\frac{a}{k}+\frac{b}{k+1})=\frac{a(k+1)+bk}{k(k+1)}$, igualando: $\frac{1}{k\cdot(k+1)}=\frac{a(k+1)+bk}{k(k+1)}$,$1=a(k+1)+bk$, hallando valores de a y b:

$\begin{align*}
k&=-1\\
1&=a(-1+1)+b(-1)\\
1&=-b\\
b&=-1\\
k&= 0\\
1&=a(0+1)+b(0)\\
1&=a
\end{align*}$

Reescribiendo nuestras fracciones : $2*(\frac{1}{k}+\frac{-1}{k+1})$, Ahora para hallar el valor de la suma tenemos que hallar el limite de la sumatoria:

$\begin{align*}
\lim_{n \to \infty} &\sum_{k=1}^{n}\Big(2*\Big(\frac{1}{k}-\frac{1}{k+1}\Big) \Big) \\ 
\lim_{n \to \infty} &2 \sum_{k=1}^{n}\Big(\Big(\frac{1}{k}-\frac{1}{k+1}\Big) \Big)\\
\end{align*}$

Luego vemos que tenemos una suma telescopica por lo anterior obtendremos:

$\begin{align*}
\lim_{n \to \infty}& 2 \cdot\Big( 1-\frac{1}{2}+\frac{1}{2}-\frac{1}{3}+\frac{1}{3}+\ldots-\frac{1}{n}+\frac{1}{n}-\frac{1}{n+1}\Big)\\
=\lim_{n \to \infty} &2\cdot\Big( 1-\frac{1}{n+1} \Big)\\
=&2\cdot (1-\frac{1}{\infty +1})\\
=&2\cdot (1-0)\\
=\underline{2}
\end{align*}$

**Punto 3.1.a**

Partiendo de que $\frac{1}{n}=\frac{1}{n+1}+\frac{1}{n\cdot(n+1)}$ operamos la parte derecha de la identidad.

$\begin{align*}
\frac{1}{n+1}+\frac{1}{n\cdot(n+1)}&=\frac{n\cdot (n+1)+(n+1)}{(n+1)\cdot n\cdot(n+1)} \\
&=\frac{(n+1)(n+1)}{(n+1)\cdot n\cdot(n+1)}\\
&=\frac{1}{n}
\end{align*}$
